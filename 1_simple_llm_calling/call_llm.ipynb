```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite")

response = llm.invoke("How many moons does Jupiter have?")
print(response.text)
```

```python
response = llm.invoke(
    [
        ["system", "You are a helpful assistant that answers in short one line"],
        ["human", "How many moons does Jupiter have?"]
    ]
)
print(response.text)
```

```python
prompt = "Write a single tagline for a quirky coffee shop. Only output the tagline, nothing else."
llm = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite", temperature=0)
response = llm.invoke(prompt)
print(response.text)
```
